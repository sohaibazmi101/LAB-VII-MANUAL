```python
import numpy as np
import matplotlib.pyplot as plt
import kagglehub

x = np.array([35,25,25,49])
labels = np.array(['Apple', 'Orange', 'Banana', 'Tomato'])
# By default start angle in 0 in degrees
plt.pie(x, labels=labels, startangle=70)
plt.show()
exp = np.array([0,0.1,0,0])
plt.pie(x, labels=labels, explode=exp, startangle=-10, shadow=True)
plt.legend(title ='Fruits')
plt.show()
```
